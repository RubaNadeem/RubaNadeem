```python
#Step 1: Upload & Load Dataset
from google.colab import files
import pandas as pd

# Upload file
uploaded = files.upload()

# Load dataset (replace with your uploaded file name if different)
df = pd.read_csv("kaggle_survey_2017_2021.csv")

# Preview first 5 rows
df.head()
```

```python
#Step 2: Explore Dataset

# Shape (rows, columns)
print("Shape:", df.shape)

# Column names
print("\nColumns:\n", df.columns.tolist())

# Summary info
df.info()

# Check missing values
print("\nMissing Values:\n", df.isnull().sum().head(20))

# Check duplicates
print("\nDuplicate rows:", df.duplicated().sum())
```

```python
#Step 3: Data Cleaning

# Drop duplicates
df = df.drop_duplicates()

# Handle missing values (example: fill with "Unknown" for categorical, median for numeric)
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].fillna("Unknown")

for col in df.select_dtypes(include=["int64","float64"]).columns:
    df[col] = df[col].fillna(df[col].median())

print("Missing values after cleaning:\n", df.isnull().sum().sum())
```

```python
#Step 4: Handle Categorical Variables

from sklearn.preprocessing import LabelEncoder

# Create a copy of dataframe so original is safe
df_encoded = df.copy()

# Initialize encoder
encoder = LabelEncoder()

# Loop through only categorical (object) columns
for col in df_encoded.select_dtypes(include="object").columns:
    try:
        # Apply encoding if the column has only string values
        df_encoded[col] = encoder.fit_transform(df_encoded[col].astype(str))
    except Exception as e:
        print(f"Skipping column: {col} because of error: {e}")

df_encoded.head()

```

```python
#Step 5: Insights Generation

# Step 5: Exploratory Data Analysis (EDA) - Updated for your dataset

# 1. Age distribution
print("\n--- Age Distribution ---")
print(df["Q1"].value_counts().head(10))

# 2. Gender distribution
print("\n--- Gender Distribution ---")
print(df["Q2"].value_counts())

# 3. Top 10 Countries
print("\n--- Top 10 Countries ---")
print(df["Q3"].value_counts().head(10))

# 4. Education Levels
print("\n--- Education Levels ---")
print(df["Q4"].value_counts())

# 5. Job Titles (Top 10)
print("\n--- Top 10 Job Titles ---")
print(df["Q5"].value_counts().head(10))

# 6. Years of Programming
print("\n--- Programming Experience ---")
print(df["Q6"].value_counts())

# 7. Languages (Q7_Part_1 ... Q7_Part_12)
print("\n--- Programming Languages Used ---")
languages = df[["Q7_Part_1","Q7_Part_2","Q7_Part_3","Q7_Part_4","Q7_Part_5",
                "Q7_Part_6","Q7_Part_7","Q7_Part_8","Q7_Part_9","Q7_Part_10",
                "Q7_Part_11","Q7_Part_12"]]

lang_counts = languages.notnull().sum().sort_values(ascending=False)
print(lang_counts)



```

```python
#Step 6: Visualizations

import matplotlib.pyplot as plt

# --- 1. Age Distribution ---
plt.figure(figsize=(8,5))
df["Q1"].value_counts().sort_index().plot(kind="bar")
plt.title("Age Distribution of Respondents")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()

# --- 2. Gender Distribution ---
plt.figure(figsize=(6,4))
df["Q2"].value_counts().plot(kind="bar", color="skyblue")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# --- 3. Top 10 Countries ---
plt.figure(figsize=(10,5))
df["Q3"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Countries of Respondents")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()

# --- 4. Education Levels ---
plt.figure(figsize=(8,5))
df["Q4"].value_counts().plot(kind="bar", color="lightgreen")
plt.title("Education Levels of Respondents")
plt.xlabel("Education Level")
plt.ylabel("Count")
plt.show()

# --- 5. Top 10 Job Titles ---
plt.figure(figsize=(10,5))
df["Q5"].value_counts().head(10).plot(kind="bar", color="orange")
plt.title("Top 10 Job Titles")
plt.xlabel("Job Title")
plt.ylabel("Count")
plt.show()

# --- 6. Years of Programming Experience ---
plt.figure(figsize=(8,5))
df["Q6"].value_counts().sort_index().plot(kind="bar", color="purple")
plt.title("Years of Programming Experience")
plt.xlabel("Experience")
plt.ylabel("Count")
plt.show()

# --- 7. Programming Languages ---
plt.figure(figsize=(10,5))
lang_counts.plot(kind="bar", color="teal")
plt.title("Programming Languages Used")
plt.xlabel("Language")
plt.ylabel("Number of Respondents")
plt.show()
```

```python
#Step 7: Final Insights Summary
# Step 7: Generate Insights (Final Summary)

print("📊 Top 5 Insights from the Kaggle Survey Data:\n")

# 1. Age Distribution
top_age = df["Q1"].value_counts().idxmax()
print(f"1️⃣ The largest age group of respondents is: {top_age}")

# 2. Gender Distribution
top_gender = df["Q2"].value_counts().idxmax()
print(f"2️⃣ Majority of respondents identify as: {top_gender}")

# 3. Country Distribution
top_country = df["Q3"].value_counts().idxmax()
print(f"3️⃣ Most respondents are from: {top_country}")

# 4. Education
top_edu = df["Q4"].value_counts().idxmax()
print(f"4️⃣ Most respondents have completed: {top_edu}")

# 5. Job Titles
top_job = df["Q5"].value_counts().idxmax()
print(f"5️⃣ The most common job title is: {top_job}")

```
